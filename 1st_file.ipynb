> ## Data Preprocessing

Load the data from csv file

```python
product_details =  pd.read_csv('product_details.csv')

# Set display options
pd.set_option('display.max_colwidth', None)  # Set max column width to None to avoid truncation
pd.set_option('display.expand_frame_repr', False)  # Prevents line wrapping of DataFrame

# Display the DataFrame
print(product_details)
```

```python
# split the product specs and othre details into diffirent columns
product_details['brand_model'] = product_details["Product Name"].str.extract(r'^(.*?)(?:\(|\|)')
product_details['specs'] = product_details["Product Name"].str.extract(r'\((.*?)\)')

# drop unwanted column 
product_details = product_details.drop('Product Name',axis=1)
product_details
```

```python

# Split the 'specs' column by '/' and create new columns
split_specs = product_details['specs'].str.split('/', expand=True)


# Assign all parts with its proper names to the columns
product_details['Processor'] = split_specs[0]
product_details['Storage'] = split_specs[1]
product_details['RAM'] = split_specs[2]
product_details['GPU'] = split_specs[2]


product_details = product_details.drop('specs',axis=1)


# Display the DataFrame
print(product_details)

```

```python
brand_split = product_details['brand_model'].str.split('-', expand=True)

product_details['Product Name'] = brand_split[0]
product_details['color'] = brand_split[1]

product_details


# product_details.to_csv('2nd_Updated_Product_details.csv', index=False, mode='w+', header=True)
```

```python

df_cleaned = product_details[product_details.count(axis=1) > 2]

df_cleaned = df_cleaned.drop('brand_model',axis=1)

print(df_cleaned)
```

```python
df_cleaned = df_cleaned.replace([None, pd.NA], 'not specified')

# df_cleaned.to_csv('cleaned_file.csv', index=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Print the DataFrame
print(df_cleaned)
```
